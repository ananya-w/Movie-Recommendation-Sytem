```python
 #strings of list ko list m badalna python function
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_only3(obj):
  l=[ ]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!= 3:
        l.append(i['name'])
        counter += 1
    else:
        break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert_only3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L

```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
 movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x]) #removes spaces to avoid clashes in recomendation sysytem
 movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x]) #removes spaces to avoid clashes in recomendation sysytem
 movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x]) #removes spaces to avoid clashes in recomendation sysytem
 movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x]) #removes spaces to avoid clashes in recomendation sysytem
```

```python
movies.head(2)
```

```python
movies['tags']=movies['overview']+movies['genres'] + movies['keywords']+movies['cast']+movies['crew']
```

```python
new_movie=movies[['movie_id','title','tags']]
```

```python
new_movie['tags']=new_movie['tags'].apply(lambda x:" ".join(x)) #it joins into a single string, with space between the elements.
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_movie['tags']=new_movie['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')#did vectorization of 5000 words and ommited stop words
```

```python
vectors=cv.fit_transform(new_movie['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity#vectore disatnce using cosine distance as it is of larger dimension
```

```python
similarity =cosine_similarity(vectors)
```

```python
similarity #its diagnol would be one as ek movie ka khud ke sath distance is 1 only
```

```python
#enumarate is used to keep a track of the indexes after sorting
#lambda isliye so that distance ke base pe sort ho
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index=new_movie[new_movie['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print("the movie id is :", i[0])
    print("the movie is :   ",new_movie.iloc[i[0]].title)
    print()
```

```python
movie_name=input("Enter the movie for which you want the recommendations: ")
recommend(movie_name)
```
